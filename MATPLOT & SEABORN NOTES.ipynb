# LINE PLOT

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

```python
# ploting the line plot
plt.plot(x,y)
plt.show()
```

```python
# Adding title and labels to the plot
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
# changing line aesthetics
plt.plot(x,y,color='r',linestyle=':',linewidth=2)
plt.show()
```

```python
plt.plot(x,y,color='r',linestyle=':',linewidth=2)
plt.title("Line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
# Adding two lines in the same plot
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x,y1,color='r',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=2)
plt.title("Line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show()
```

```python
# Adding sub-plots
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x

plt.subplot(1,2,1)
plt.plot(x,y1,color='r',linestyle=':',linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=2)

plt.show()
```

# BAR PLOT

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
student = {"bob":87,"matt":56,"sam":27}
```

```python
names = list(student.keys())
values = list(student.values())
```

```python
plt.bar(names,values)
plt.show()
```

```python
# Adding titles and labels
plt.bar(names,values)
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
# Horizontl bar plot
plt.barh(names,values,color='g')# barh = horizontal
plt.title('Barplot')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

# SCATTER PLOT

```python
# creating a basic scatter-plot
x = [10,20,30,40,50,60,70,80,90]
y = [8,1,7,2,0,3,7,3,2]

plt.scatter(x,y)
plt.show()
```

```python
# changing marks aesthetics
x = [10,20,30,40,50,60,70,80,90]
y = [8,1,7,2,0,3,7,3,2]
plt.scatter(x,y,marker="*",c="g",s=100) # marker is the style and s is the size
plt.show()
```

```python
# adding two markers in the same plot
x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
b = [7,2,5,6,9,1,4,5,3]
plt.scatter(x,a,marker="*",c="g",s=100)
plt.scatter(x,b,marker=".",c="r",s=100)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
b = [7,2,5,6,9,1,4,5,3]
plt.xlabel("a part")
plt.ylabel("b part")
plt.scatter(x,a,marker="*",c="g",s=100)
plt.scatter(x,b,marker=".",c="r",s=100)
plt.show()
```

```python
# adding sub plot
x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
b = [7,2,5,6,9,1,4,5,3]

plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c="g",s=100)

plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c="r",s=100)

plt.show()
```

# PIE -CHART

```python
import numpy as np
from matplotlib import pyplot as plt

# creating data
fruit = ['apple','orange','mango','guava']
quantity = [67,34,100,29]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
# changing the aesthetics

#autopct for percentage 
plt.pie(quantity,labels = fruit,autopct='%0.1f%%',colors=['yellow','grey','blue','green'])
plt.show()
```
